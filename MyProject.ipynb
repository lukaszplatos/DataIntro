```python
# Looping
for lab, row in top.iterrows():
    if "burger" in str(row['Segment_Category']).lower():
        top.loc[lab, 'Burger'] = 'OK'
    else:
        top.loc[lab, 'Burger'] = 'NO OK'
top.head(20)        
```

```python
for lab, row in top.iterrows():
    if "mexican" in str(row['Segment_Category']).lower():
        top.loc[lab, 'Mexican'] = 'OK'
    else:
        top.loc[lab, 'Mexican'] = 'NO OK'
top.head(20)        
```
